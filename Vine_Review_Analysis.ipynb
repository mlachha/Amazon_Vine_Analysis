```python
#Create a pandas dataframe from the vine CSV
vine_df = pd.read_csv('vine_data.csv')
vine_df
```

```python
# filter for only columns with 20 or more total votes 
over_20_votes_df = vine_df[(vine_df.total_votes >=20)]
over_20_votes_df
```

```python
#filter for the most helpful votes
helpful_votes_df = over_20_votes_df.loc[(over_20_votes_df['helpful_votes'] /over_20_votes_df['total_votes'] >= 0.5)]
helpful_votes_df
```

```python
#filter for the votes that are part of the PAID vine program 
part_of_vine_df = helpful_votes_df.loc[helpful_votes_df.vine == 'Y']
part_of_vine_df
```

```python
#filter for the votes that are part of the PAID vine program 
not_part_of_vine_df = helpful_votes_df.loc[helpful_votes_df.vine == 'N']
not_part_of_vine_df
```

```python
# Determine total number of reviews
total_reviews =helpful_votes_df.review_id.count()
total_reviews
```

```python
# Determine total number of paid reviews
vine_reviews = part_of_vine_df.review_id.count()
vine_reviews
```

```python
# Determine total number of unpaid reviews
not_vine_reviews = not_part_of_vine_df.review_id.count()
not_vine_reviews
```

```python
# Determine number of 5-star reviews
five_star = helpful_votes_df[(helpful_votes_df.star_rating == 5 )].review_id.count()
five_star
```

```python
# Determine number of 5-star paid reviews
five_paid = part_of_vine_df[(part_of_vine_df.star_rating == 5 )].review_id.count()
five_paid
```

```python
# Determine number of 5-star unpaid reviews
five_unpaid = not_part_of_vine_df[(not_part_of_vine_df.star_rating == 5 )].review_id.count()
five_unpaid
```

```python
# Determine percentage of 5-star paid reviews
#in this example there is no paid reviews as we proved before but we can still calculate that:
paid_percentage = vine_reviews/total_reviews * 100
paid_percentage 
```

```python
# Determine percentage of 5-star unpaid reviews
unpaid_percentage = not_vine_reviews/total_reviews * 100
unpaid_percentage 
```
